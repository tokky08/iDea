```python
# ブラウザを操作するオブジェクトの生成
driver = webdriver.Chrome("driver/chromedriver")
# 指定したURLへの移動
driver.get("https://google.co.jp")

value = input()

# 検索窓にSeleniumと入力する
selector = '#tsf > div:nth-child(2) > div.A8SBwf > div.RNNXgb > div > div.a4bIc > input'
element = driver.find_element_by_css_selector(selector)
element.send_keys(value)

# enterキーを押す
element.send_keys(Keys.ENTER)


# ページのHTMLを取得
html = driver.page_source
# ブラウザの終了
driver.close()

# 取得したHTMLからBeautifulSoupオブフェクトを生成
# scriptやstyle及びその他タグの除去
soup = BeautifulSoup(html, "lxml")
for s in soup(['script', 'style']):
    s.decompose()

text = ' '.join(soup.stripped_strings)
```

```python
m = MeCab.Tagger("")

nouns = [line for line in m.parse(text).splitlines()
               if "名詞" in line.split()[-1]]

relation_list = []
for str in nouns:
#     print(str.split())
    relation_list.append(str.split()[0])
print(relation_list)
```
